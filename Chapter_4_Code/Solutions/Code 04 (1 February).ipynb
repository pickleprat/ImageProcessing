```python
from skimage.data import camera
I = camera()
print(I.shape)
```

```python
from matplotlib.pyplot import imshow
imshow(I, cmap='gray')
```

```python
print(I)
```

```python
from numpy import transpose
A = transpose(I)
print(A)
```

```python
imshow(A, cmap='gray')
```

```python
from numpy import flipud, fliplr, rot90

B = flipud(I)
print(B)
```

```python
imshow(B, cmap='gray')
```

```python
C = fliplr(I)
print(C)
```

```python
imshow(C, cmap='gray')
```

```python
D = rot90(I)
imshow(D, cmap='gray')
```

```python
E = rot90(rot90((I)))
imshow(E, cmap='gray')
```

```python
from matplotlib.pyplot import subplot

subplot(2,3,1)
imshow(I, cmap='gray')

subplot(2,3,2)
imshow(A, cmap='gray')

subplot(2,3,3)
imshow(B, cmap='gray')

subplot(2,3,4)
imshow(C, cmap='gray')

subplot(2,3,5)
imshow(D, cmap='gray')

subplot(2,3,6)
imshow(E, cmap='gray')
```

```python
from skimage.io import imread

I1 = imread('image_1.bmp')
I2 = imread('image_2.bmp')
```

```python
print(I1.shape, I2.shape)
```

```python
print(I1.max(), I2.max())
```

```python
print(I1.min(), I2.min())
```

```python
a = 0
b = 255
print(a and a)
print(a and b)
print(b and a)
print(b and b)
```

```python
print(a or a)
print(a or b)
print(b or a)
print(b or b)
```

```python
from numpy import zeros

M = I1.shape[0]
N = I2.shape[1]

B = zeros((M,N), dtype='uint8')
C = zeros((M,N), dtype='uint8')
D = zeros((M,N), dtype='uint8')
E = zeros((M,N), dtype='uint8')

for i in range(M):
    for j in range(N):
        B[i][j] = not I1[i][j]
        C[i][j] = not I2[i][j]
        D[i][j] = I1[i][j] and I2[i][j]
        E[i][j] = I1[i][j] or I2[i][j]
        
subplot(2,3,1)
imshow(I1, cmap='gray')

subplot(2,3,2)
imshow(I2, cmap='gray')

subplot(2,3,3)
imshow(B, cmap='gray')

subplot(2,3,4)
imshow(C, cmap='gray')

subplot(2,3,5)
imshow(D, cmap='gray')

subplot(2,3,6)
imshow(E, cmap='gray')
```
