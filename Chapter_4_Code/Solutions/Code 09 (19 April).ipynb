```python
from skimage.io import imread

f = imread('Image5.tif', as_gray=True)
```

```python
print(f.shape)
print(f.max())
print(f.min())
```

```python
from matplotlib.pyplot import imshow, show

imshow(f, cmap='gray')
show()
```

# Roberts

```python
from skimage.filters import roberts
from matplotlib.pyplot import title

g = roberts(f)
imshow(g, cmap='gray')
title('Edge detection using Roberts operator')
show()
```

# Prewitt

```python
from skimage.filters import prewitt, prewitt_v, prewitt_h

g1 = prewitt(f)
imshow(g1, cmap='gray')
title('Edge detection using Prewitt opertor')
show()
```

```python
g2 = prewitt_h(f)
imshow(g2, cmap='gray')
title('Edge detection using Prewitt horizontal opertor')
show()
```

```python
g3 = prewitt_v(f)
imshow(g3, cmap='gray')
title('Edge detection using Prewittvertical opertor')
show()
```

# Sobel

```python
from skimage.filters import sobel, sobel_h, sobel_v

g1 = sobel(f)
g2 = sobel_h(f)
g3 = sobel_v(f)
```

```python
imshow(g1, cmap='gray')
title('Edge detection using Sobel operator')
show()
```

```python
imshow(g2, cmap='gray')
title('Edge detection using Sobel horizontal operator')
show()
```

```python
imshow(g3, cmap='gray')
title('Edge detection using Sobel vertical operator')
show()
```
