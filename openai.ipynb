```python
import os 
import openai 
import cv2 
import numpy as np 
from matplotlib import pyplot as plt 

openai.api_key = os.environ['OPEN_AI_API']
img = cv2.resize(cv2.imread('static/fari_beautiful.jpeg'), (1024, 1024)) 
mask = img.copy()
mask[340:440, 340:490] = 0
mask[350:450, 610:780] = 0


img = cv2.cvtColor(img, cv2.COLOR_RGB2RGBA)
mask = cv2.cvtColor(mask, cv2.COLOR_RGB2RGBA)
mask = cv2.cvtColor(mask, cv2.COLOR_RGBA2GRAY)


_, buffer = cv2.imencode('.png', img)
image = buffer.tobytes()

_, buffer_mask = cv2.imencode('.png', mask)
image_mask = buffer_mask.tobytes()

response = openai.Image.create_edit(
  prompt="Wooden plank",
  image=image,
  mask = image_mask,
  n=10,
  size="1024x1024", 

)

response
```
