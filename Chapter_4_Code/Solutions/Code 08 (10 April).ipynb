```python
from skimage.io import imread

f1 = imread('Image1.tif')
f2 = imread('Image2.tif')
f3 = imread('Image3.tif')
```

```python
from matplotlib.pyplot import imshow, show

imshow(f1, cmap='gray')
print('Size of image:',f1.shape)
show()
```

```python
a = f1[50,:]
print(a)
```

```python
from matplotlib.pyplot import plot, title, xlabel, ylabel

plot(a)
title('Step edge')
xlabel('Column index')
ylabel('Graylevels')
show()
```

```python
print('Size of image:',f2.shape)
imshow(f2, cmap='gray')
show()
```

```python
b = f2[60,:]
print(b.shape)
print(b)
```

```python
plot(b)
xlabel('Column index')
ylabel('Graylevel')
title('Ramp edge')
show()
```

```python
print('Size of image:',f3.shape)
imshow(f3, cmap='gray')
show()
```

```python
c = f3[100,:]
print('Number of elements in array c is: ',len(c))
print(c)
```

```python
plot(c)
xlabel('Column index')
ylabel('Graylevel')
title('Roof edge')
show()
```

```python
f4 = imread('Image4.tif')
print('Size of image:', f4.shape)
imshow(f4, cmap='gray')
show()
```

```python
d = f4[150,:]
print('Number of element is d is ',len(d))
print(d)

plot(d)
xlabel('Column index')
ylabel('Intensties')
title('Plotting particular row')
show()
```

```python
def first_derivative(input):
    output = [0]*len(input)
    for i in range(1, len(input)):
        output[i] = input[i] - input[i-1]
    return output
```

```python
da = first_derivative(a)

plot(da)
xlabel('Column index')
ylabel('First derivatives of graylevel')
title('First derivative of Step edge')
show()
```

```python
db  = first_derivative(b)

plot(db)
xlabel('Column index')
ylabel('First derivatives of graylevel')
title('First derivative of ramp edge')
show()
```

```python
dc  = first_derivative(c)

plot(dc)
xlabel('Column index')
ylabel('First derivatives of graylevel')
title('First derivative of Roof Edge')
show()
```

```python
dd  = first_derivative(d)

plot(dd)
xlabel('Column index')
ylabel('First derivatives of graylevel')
title('First derivative of Circle image')
show()
```

```python
d2a = first_derivative(first_derivative(a))

plot(d2a)
xlabel('Column index')
ylabel('Second derivatives of graylevel')
title('Second derivative of Step edge')
show()
```

```python
d2b = first_derivative(first_derivative(b))

plot(d2b)
xlabel('Column index')
ylabel('Second derivatives of graylevel')
title('Second derivative of Ramp edge')
show()
```

```python
d2c = first_derivative(first_derivative(c))

plot(d2c)
xlabel('Column index')
ylabel('Second derivatives of graylevel')
title('Second derivative of Roof edge')
show()
```

```python
d2d = first_derivative(first_derivative(d))

plot(d2d)
xlabel('Column index')
ylabel('Second derivatives of graylevel')
title('Second derivative of Circle image')
show()
```
