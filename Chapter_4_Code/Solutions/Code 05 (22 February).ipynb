```python
from skimage.data import camera
from matplotlib.pyplot import imshow
Image = camera()
imshow(Image, cmap='gray')
```

Log Transformation
$ s = c \log(1+r) $

```python
from numpy import zeros, log

M = Image.shape[0]
N = Image.shape[1]
output_image_1 = zeros((M,N))

c = 1

for i in range(M):
    for j in range(N):
        output_image_1[i][j] = c * log (1+Image[i][j])
imshow(output_image_1, cmap='gray')
```

Power-law (Gamma) Transformation 
$ s = c  r^\gamma$

```python
c = 1
gamma = 0.1

output_image_2 = zeros((M,N))

for i in range(M):
    for j in range(N):
        output_image_2[i][j] = c * (Image[i][j])**gamma

imshow(output_image_2, cmap='gray')
```

```python
c = 1
gamma = 3

output_image_3 = zeros((M,N))

for i in range(M):
    for j in range(N):
        output_image_3[i][j] = c * (Image[i][j])**gamma

imshow(output_image_3, cmap='gray')
```

Image Histogram

```python
from scipy.ndimage.measurements import histogram
from matplotlib.pyplot import bar, xlabel, ylabel

h1 = histogram(Image, Image.min(), Image.max(), 255)

bar(range(255), h1)
xlabel('Graylevels')
ylabel('Frequency')
```

```python
h2 = histogram(output_image_1, output_image_1.min(), output_image_1.max(), 255)

bar(range(255), h2)
xlabel('Graylevels')
ylabel('Frequency')
```

```python
h3 = histogram(output_image_2, output_image_2.min(), output_image_2.max(), 255)

bar(range(255), h3)
xlabel('Graylevels')
ylabel('Frequency')
```

```python
h4 = histogram(output_image_3, output_image_3.min(), output_image_3.max(), 255)

bar(range(255), h4)
xlabel('Graylevels')
ylabel('Frequency')
```

```python
from matplotlib.pyplot import subplot
subplot(1,2,1)
imshow(Image, cmap='gray')
subplot(1,2,2)
bar(range(255), h1)
```

Smoothing and Shapering using filters

```python
from skimage.data import coins
A = coins()
imshow(A, cmap='gray')
```

Box filter

```python
from scipy.ndimage import correlate
from numpy import ones

filter = (1/81)*ones((9,9))
filter
```

```python
filtered_image = correlate(A, filter)
imshow(filtered_image, cmap='gray')
```

Gaussian Filter

```python
from skimage.filters import gaussian

filtered_image_1 = gaussian(A, sigma=10)
imshow(filtered_image_1, cmap='gray')
```

Laplace filter

```python
from skimage.data import clock

B = clock()
imshow(B, cmap='gray')
```

```python
kernel = [[0, 1, 0], 
          [1, -4, 1],
          [0, 1, 0]]
sharp_image_1 = correlate(B, kernel)
imshow(sharp_image_1, cmap='gray')
```

```python
kernel = [[1, 1, 1], 
          [1, -8, 1],
          [1, 1, 1]]
sharp_image_2 = correlate(B, kernel)
imshow(sharp_image_2, cmap='gray')
```

Roberts cross gradient operator

```python
kernel = [[-1, 0], 
          [0, 1]]
sharp_image_3 = correlate(B, kernel)
imshow(sharp_image_3, cmap='gray')
```

```python
kernel = [[0, -1], 
          [1, 0]]
sharp_image_4 = correlate(B, kernel)
imshow(sharp_image_4, cmap='gray')
```

Sobel filter

```python
from skimage.filters import sobel, sobel_h, sobel_v

sobel_image_1 = sobel(B)
imshow(sobel_image_1, cmap='gray')
```

```python
sobel_image_2 = sobel_h(B)
imshow(sobel_image_2, cmap='gray')
```

```python
sobel_image_3 = sobel_v(B)
imshow(sobel_image_3, cmap='gray')
```
