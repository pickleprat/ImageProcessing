```python
from numpy.fft import fft, ifft
from numpy import array

A = array([1,2,3,4])
print('The given array is',A)

B = fft(A)
print('The Fourier transform of above array is',B)

C = ifft(B)
print('The inverse Fourier transform is',C)
```

```python
from numpy.fft import fft2, ifft2

D = array([[1,0,1,1],[1,1,-2,1],[0,1,1,2],[0,0,2,1]])
print('The matrix is\n',D)

E = fft2(D)
print('The Fourier transform of above matrix is\n',E)

F = abs(E)
print('Absolute value of Fourier transform is\n',F)

G = ifft2(E)
print('Invsrse Fourier transform is\n',G)
```

```python
from skimage.data import coins
from matplotlib.pyplot import imshow

image = coins()
imshow(image, cmap='gray')
```

```python
from skimage.filters import butterworth

I1 = butterworth(image, cutoff_frequency_ratio=0.02, order=3, high_pass=False)
imshow(I1, cmap='gray')
```

```python
I2 = butterworth(image, cutoff_frequency_ratio=0.08, order=3, high_pass=False)
imshow(I2, cmap='gray')
```

```python
I3 = butterworth(image, cutoff_frequency_ratio=0.16, order=3, high_pass=False)
imshow(I3, cmap='gray')
```

```python
I4 = butterworth(image, cutoff_frequency_ratio=0.02, order=3, high_pass=True)
imshow(I4, cmap='gray')
```

```python
I5 = butterworth(image, cutoff_frequency_ratio=0.08, order=3, high_pass=True)
imshow(I5, cmap='gray')
```
